```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n-seg.pt")  # load a pretrained model (recommended for training)
```

```python
# Freeze all layers
for param in model.model.parameters():
    param.requires_grad = False

# Unfreeze last 8 layers
for layer in model.model.model[-15:]:
    for param in layer.parameters():
        param.requires_grad = True

print("âœ… Frozen all layers except last 8")
```

```python
model.train(
    data="/content/Road_crack_detection-4/data.yaml",
    task="segment",

    epochs=150,
    imgsz=768,
    batch=8,
    device=0,
    workers=4,

    # Fine-tuning hyperparameters
    lr0=5e-5,          # low LR for fine-tuning
    optimizer="AdamW",
    pretrained=True,

    # Augmentations (good for cracks)
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=0.0,
    flipud=0.0,
    fliplr=0.5,

    patience=20,
    save=True
)
```

```python
# @title
model.val()
```
